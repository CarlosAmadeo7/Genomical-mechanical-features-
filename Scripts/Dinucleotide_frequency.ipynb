## Loading libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Loading data 

```python
data=pd.read_csv('C:/Users/amade/Desktop/testing.csv', header=None)
data.head(5)
```

### Determining the parameters for plotting 

```python
dinucleotide_name = data[0]  # dinucleotide names. 
first_genome = data[1]  # dinucleotide frequency of the first genome.
second_genome = data[2]  # dinucleotide frequency of the second genome.
bar_width = 0.35
x = range(len(dinucleotide_name))  # to be used as length axis of the plot
# Label of the axes
xlabs = ["AA", "AC", "AG", "AT",
         "CA", "CC", "CG", "CT","GA", "GC", "GG", "GT","TA", "TC", "TG", "TT"]
ylabs = ["AA", "AC", "AG", "AT",
         "CA", "CC", "CG", "CT","GA", "GC", "GG", "GT","TA", "TC", "TG", "TT"]
```

### Plotting 

```python
fig, ax = plt.subplots(figsize=(10,5))  # Setting the size of the figure 
plt.bar(x, first_genome, width=bar_width, label='First genome', color='firebrick',linewidth=2,edgecolor='black')
plt.bar([i + bar_width for i in x], second_genome, width=bar_width, label='Second genome', color='mediumblue',linewidth=2,edgecolor='black')

plt.xticks([i + bar_width / 2 for i in x], dinucleotide_name, rotation=0, ha="center")  
ax.set_xticks(np.arange(len(xlabs)), labels = xlabs)
ax.tick_params(axis='both', labelsize=14) 

# Add labels for the y-axis and title
plt.xlabel('Dinucleotide',size=16)  # label of the x axis 
plt.ylabel('Frequency',size=16)  # label of the y axis 
plt.title('Dinucleotide frequency of the first & second genome',size=14,x=0.5, y=1.05)

# Add a legend to distinguish the files
plt.legend(fontsize="13")
plt.show()
#fig.savefig('C:/Users/amade/Documents/UofSC/Bioinformatics/Github/Article/Figures/Dinucleotide_frequency.jpg', bbox_inches='tight',dpi=150)
```
