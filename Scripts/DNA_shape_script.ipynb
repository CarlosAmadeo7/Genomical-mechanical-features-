## DNA shape plotting: heatmaps of structural features vs location in the genome

### Helical twist

```python
HELT_column= df.loc[:,"HELT"] # Extracting the Helt column from the dataframe
helt = HELT_column.to_numpy() # converting the values to a numpy 
a = np.expand_dims(helt, axis=0) # changing dimensions of the orginal column to a single dimension 
print(a.shape)
```

```python
fig, ax = plt.subplots(figsize=(16,1.7)) # width and height in inches, #reading image 
im = ax.imshow(a,origin='lower',aspect='auto',interpolation='nearest',cmap="viridis",vmin=30,vmax=38)
ax.set_yticks([])    #it hides the y-axis labels
plt.ylabel('HelT degree',size=14)
plt.title('Helical Twist (HelT)',size=18)
# Locating current axes
divider = make_axes_locatable(ax)
colorbar_axes = divider.append_axes("right",size="3%",pad=0.05)
# Using new axes for colorbar
plt.colorbar(im, cax=colorbar_axes)
#plt.savefig("C:/Users/amade/Documents/UofSC/Bioinformatics/Github/Article/Figures/DNA_shape_Helical_Twist.png",format="png", bbox_inches='tight')
plt.show()
```

### Minor groove width

```python
MGW_column= df.loc[:,"MGW"]   # Extracting the Helt column from the dataframe
mgw = MGW_column.to_numpy() # converting the values to a numpy 
b = np.expand_dims(mgw, axis=0) # changing dimensions of the orginal column to a single dimension 
print(b.shape)
```

```python
fig, ax = plt.subplots(figsize=(16,1.7)) # width and height in inches
im = ax.imshow(b, origin='lower',aspect='auto',interpolation='nearest',cmap="binary")
ax.set_yticks([])
plt.ylabel('MGW degree',size=14)
plt.title('Minor Groove width (MGW)',size=18)
# Locating current axes
divider = make_axes_locatable(ax)
colorbar_axes = divider.append_axes("right",size="3%",pad=0.05)
# Using new axes for colorbar
plt.colorbar(im, cax=colorbar_axes)
#plt.savefig("C:/Users/amade/Documents/UofSC/Bioinformatics/Github/Article/Figures/DNA_shape_Minor_groove_width.png" ,format="png", bbox_inches='tight')
plt.show()
```

### Propeller twist 

```python
PROT_column= df.loc[:,"PROT"]  # Extracting the Helt column from the dataframe
prot = PROT_column.to_numpy()  # Converting the values to a numpy 
c = np.expand_dims(prot, axis=0)   # Changing dimensions of the orginal column to a single dimension 
print(c.shape)
```

```python
fig, ax = plt.subplots(figsize=(16,1.7)) # width and height in inches
im = ax.imshow(c, origin='lower',aspect='auto',interpolation='nearest',cmap="spring")
ax.set_yticks([])#it hides the y-axis labels
plt.ylabel('ProT degree',size=14)
plt.title('Propeller Twist (ProT)',size=18)
#cbar = fig.colorbar(im)
# Locating current axes
divider = make_axes_locatable(ax)
colorbar_axes = divider.append_axes("right",size="3%",pad=0.05)
# Using new axes for colorbar
plt.colorbar(im, cax=colorbar_axes)
#plt.savefig("C:/Users/amade/Documents/UofSC/Bioinformatics/Github/Article/Figures/DNA_shape_Propeller_Twist.png",format="png", bbox_inches='tight')
plt.show()
```

### Roll

```python
Roll_column= df.loc[:,"ROLL"]   # Extracting the Helt column from the dataframe
roll = Roll_column.to_numpy()   # Converting the values to a numpy 
d = np.expand_dims(roll, axis=0)   # Changing dimensions of the orginal column to a single dimension 
print(d.shape)
```

```python
fig, ax = plt.subplots(figsize=(16,1.7)) # width and height in inches
im = ax.imshow(d, origin='lower',aspect='auto',interpolation='nearest',cmap="winter")
ax.set_yticks([])#it hides the y-axis labels
plt.ylabel('Roll degree',size=14)
plt.title('Roll (roll)',size=18)
#cbar = fig.colorbar(im)
# Locating current axes
divider = make_axes_locatable(ax)
colorbar_axes = divider.append_axes("right",size="3%",pad=0.05)
# Using new axes for colorbar
plt.colorbar(im, cax=colorbar_axes)
#plt.savefig("C:/Users/amade/Documents/UofSC/Bioinformatics/Github/Article/Figures/DNA_shape_Roll.png",format="png", bbox_inches='tight')
plt.show()
```

## Correlation plot of the genome

```python
correlation= df.corr()  #Correlation of DNA shape
correlation
```

```python
plt.figure(figsize=(6,4))
corre_heatmap=sns.heatmap(correlation,cmap="RdBu",annot=True,linewidth=1,linecolor="Black",annot_kws={'size': 14}) 
corre_heatmap.set_xticklabels(corre_heatmap.get_xticklabels(), fontsize=12)
corre_heatmap.set_yticklabels(corre_heatmap.get_yticklabels(), fontsize=12)
plt.title("DNA shape correlation of PsgM04 genome")
plt.show()
```
