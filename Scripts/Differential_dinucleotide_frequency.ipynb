## Loading libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Loading and finding dinucleotide frequency of the first genome

```python
freq_list=[]  # Creating an empty array 
with open("C:/Users/amade/Documents/DNA_Shape/nonheaders/M02_predictions/M02_phage_genome.txt", 'r') as ins:
    array=[]   # temporary list 
    for line in ins:
        array.append(line.replace('\n',''))    # appending the frequency of each dinucloetide in the initial array 
    freq_list.append((array[0].count('AA')/len(array[0]))*100)  # Finding dinucleotide frequency 
    freq_list.append((array[0].count('AC')/len(array[0]))*100)  
    freq_list.append((array[0].count('AG')/len(array[0]))*100)  
    freq_list.append((array[0].count('AT')/len(array[0]))*100) 
    freq_list.append((array[0].count('CA')/len(array[0]))*100)
    freq_list.append((array[0].count('CC')/len(array[0]))*100)
    freq_list.append((array[0].count('CG')/len(array[0]))*100)
    freq_list.append((array[0].count('CT')/len(array[0]))*100)
    freq_list.append((array[0].count('GA')/len(array[0]))*100)
    freq_list.append((array[0].count('GC')/len(array[0]))*100)
    freq_list.append((array[0].count('GG')/len(array[0]))*100)
    freq_list.append((array[0].count('GT')/len(array[0]))*100)
    freq_list.append((array[0].count('TA')/len(array[0]))*100)
    freq_list.append((array[0].count('TC')/len(array[0]))*100)
    freq_list.append((array[0].count('TG')/len(array[0]))*100)
    freq_list.append((array[0].count('TT')/len(array[0]))*100)
    
print(freq_list)   # Dinucleotide frequency in percentage 
```

### Loading and finding dinucleotide frequency of the second genome

```python
freq_list1=[]   # Creating an empty array 
with open("C:/Users/amade/Documents/DNA_Shape/nonheaders/M04_predictions/M04_phage_genome.txt", 'r') as ins:
    array=[]    # temporary list 
    for line in ins:
        array.append(line.replace('\n',''))      # appending the frequency of each dinucloetide in the initial array 
    freq_list1.append((array[0].count('AA')/len(array[0]))*100)  # Finding dinucleotide frequency 
    freq_list1.append((array[0].count('AC')/len(array[0]))*100)  
    freq_list1.append((array[0].count('AG')/len(array[0]))*100)  
    freq_list1.append((array[0].count('AT')/len(array[0]))*100) 
    freq_list1.append((array[0].count('CA')/len(array[0]))*100)
    freq_list1.append((array[0].count('CC')/len(array[0]))*100)
    freq_list1.append((array[0].count('CG')/len(array[0]))*100)
    freq_list1.append((array[0].count('CT')/len(array[0]))*100)
    freq_list1.append((array[0].count('GA')/len(array[0]))*100)
    freq_list1.append((array[0].count('GC')/len(array[0]))*100)
    freq_list1.append((array[0].count('GG')/len(array[0]))*100)
    freq_list1.append((array[0].count('GT')/len(array[0]))*100)
    freq_list1.append((array[0].count('TA')/len(array[0]))*100)
    freq_list1.append((array[0].count('TC')/len(array[0]))*100)
    freq_list1.append((array[0].count('TG')/len(array[0]))*100)
    freq_list1.append((array[0].count('TT')/len(array[0]))*100)
    
print(freq_list1)   # Dinucleotide frequency in percentage 
```

### Finding the differential dinucleotide frequency bettween the genomes

```python
difference=[(x-y)/100 for x in freq_list for y in freq_list1]    # finding the difference bettween the genomes
differences_2d = np.array(difference).reshape(len(freq_list1), -1)   # reshaping the array 
df = pd.DataFrame(differences_2d, columns=[f'Difference_{i+1}' for i in range(len(freq_list1))])  # Converting to a dataframe
df
```

## Plotting 

```python
#Labels of the axes
xlabs = ["AA", "AC", "AG", "AT",
         "CA", "CC", "CG", "CT","GA", "GC", "GG", "GT","TA", "TC", "TG", "TT"]
ylabs = ["AA", "AC", "AG", "AT",
         "CA", "CC", "CG", "CT","GA", "GC", "GG", "GT","TA", "TC", "TG", "TT"]
```

```python
fig, ax = plt.subplots(figsize=(6,6))  # setting the size of the figure 
im=ax.imshow(df,cmap='coolwarm')

ax.set_xticks(np.arange(len(xlabs)), labels = xlabs)  # setting the x labels 
ax.set_yticks(np.arange(len(ylabs)), labels = ylabs)  # setting the y labels
ax.tick_params(axis='both', labelsize=12) 

plt.setp(ax.get_xticklabels(), rotation = 90,ha = "right", rotation_mode = "anchor")

plt.title('Differential Dinucleotide frequency bettween first & second genome', size=14,  x=0.55, y=1.05)
plt.xlabel('First',size=14,x=0.5, y=10)
plt.ylabel('Second',size=13,x=-1, y=0.5)

cbar = ax.figure.colorbar(im, ax = ax,fraction=0.046, pad=0.04)
#fig.savefig('C:/Users/amade/Documents/UofSC/Bioinformatics/Github/Article/Figures/Differential_frequency.jpg', bbox_inches='tight',dpi=150)
plt.show() 
```
