## Loading libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import re
import seaborn as sns
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

### Correlation analysis

### Loading data

```python
M04_MP=pd.read_csv("C:/Users/amade/Documents/DNA_Shape/nonheaders/M04_predictions/Mechanical_properties.csv") # Loading the csv file
df = pd.DataFrame(M04_MP)  # Dataframe
print(df)
```

### Finding pairwise correlation bettween the DNA structural features

```python
correlation= df.corr()  
correlation
```

```python
plt.figure(figsize=(6,4))
corre_heatmap=sns.heatmap(correlation,cmap="RdBu",annot=True,linewidth=1,linecolor="Black",annot_kws={'size': 14}) 
corre_heatmap.set_xticklabels(corre_heatmap.get_xticklabels(), fontsize=12)
corre_heatmap.set_yticklabels(corre_heatmap.get_yticklabels(), fontsize=12)
plt.title("DNA shape correlation of PsgM04 genome")
plt.savefig("C:/Users/amade/Documents/UofSC/Bioinformatics/Github/Article/Figures/DNA_shape_Correlation.png",format="png", bbox_inches='tight')
plt.show()
```
